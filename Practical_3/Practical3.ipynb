```python
import pandas as pd
df = pd.read_csv("teacher_ratings.csv")

print(df.head())
```

Q1. Identify all duplicate cases using prof. Using all observations, find the average and standard deviation for age. Repeat the analysis by first filtering the dataset to include one observation per instructor (94 observations).

A1.

```python
# Check for duplicates
df.duplicated(subset='prof').sum()

# Mean and standard deviation for age
mean_all = df['age'].mean()
std_all = df['age'].std()

# Remove duplicates 
df_unique = df.drop_duplicates(subset='prof')

# Mean and standard deviation after filtering
mean_unique = df_unique['age'].mean()
std_unique = df_unique['age'].std()

print("All Observations → Mean:", mean_all, "Std Dev:", std_all)
print("Unique Professors → Mean:", mean_unique, "Std Dev:", std_unique)
```

Q3. Plot the relationship between age and teaching evaluation scores.

A3.

```python
import matplotlib.pyplot as plt

plt.scatter(df['age'], df['eval'], color='mediumorchid', alpha=0.6)
plt.title('Relationship Between Age and Teaching Evaluation Scores')
plt.xlabel('Instructor Age')
plt.ylabel('Evaluation Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Q4. Using gender-differentiated scatter plots, plot the relationship between age and teaching evaluation scores.

A4.

```python
import seaborn as sns

sns.scatterplot(x='age', y='eval', hue='gender', data=df, palette='Set2')
plt.title('Age vs Evaluation Scores by Gender')
plt.xlabel('Instructor Age')
plt.ylabel('Evaluation Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Q5. Create a scatterplot of age and evaluation scores, differentiated by gender and tenure.

A5.

```python
sns.scatterplot(
    x='age', y='eval', hue='gender', style='tenure', 
    data=df, palette='Set1'
)
plt.title('Age vs Evaluation Scores by Gender and Tenure')
plt.xlabel('Instructor Age')
plt.ylabel('Evaluation Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
